```python
#Imports library
import random

#Function that generates random number
def Random_Number_generator(x, y):
    Random_number = random.randint(x, y)
    return Random_number


#Main program
print("Welcome to Miki Number Guessing Game")
print("Enter the number against the levels to select a level.")
print("1. Easy")
print("2. Medium")
print("3. Hard")

#Prompts user to select a level
level = int(input("Choose a Level: "))
            
            
if level == 1:
    print("*** Level 1 ***")
    print("In this level, you are expected to guess a number between 1 and 10. You have 6 chances. Goodluck!")
    #Generates a random number and stores same in the Random_Number variable
    Random_Number=Random_Number_generator(1,10)
    guess_count =0
    guess_limit = 6
    #Executes provided user has not  exceeded the guess_limit 
    while guess_count < guess_limit:
        guess=int(input("Make a Guess: "))
        guess_count +=1
       #Compares user's input with the random number
        if (guess == Random_Number):
            print("You got it right!")
            break
        else:
            print("That was wrong!")
            print("You have used " + str(guess_count) + " of 6 chances")
    else:
        print("Game Over!")
elif level==2:
    print("*** Level 2 ***")
    print("In this level, you are expected to guess a number between 1 and 20. You have 4 chances. Goodluck!")
    #Generates a random number and stores same in the Random_Number variable
    Random_Number=Random_Number_generator(1,20)
    guess_count =0
    guess_limit = 4
    #Executes provided user has not  exceeded the guess_limit
    while guess_count < guess_limit:
        guess=int(input("Make a Guess: "))
        guess_count +=1
        if guess == Random_Number:
            print("You got it right!")
            break
        else:
            print("That was wrong!")
            print("You have used " + str(guess_count) + " of 4 chances")
    else:
        print("Game Over!")
elif level ==3:
    print("*** Level 3 ***")
    print("In this level, you are expected to guess a number between 1 and 50. You have 3 chances. Goodluck!")
    #Generates a random number and stores same in the Random_Number variable
    Random_Number=Random_Number_generator(1,50)
    guess_count =0
    guess_limit = 3
    #Executes provided user has not  exceeded the guess_limit
    while guess_count < guess_limit:
        guess=int(input("Make a Guess: "))
        guess_count +=1
         #Compares user's input with the random number
        if guess == Random_Number:
            print("You got it right!")
            break
        else:
            print("That was wrong!")
            print("You have used " + str(guess_count) + " of 3 chances")
    else:
        print("Game Over!")
```
